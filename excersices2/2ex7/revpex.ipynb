Given: A DNA string of length at most 1 kbp in FASTA format.

Return: The position and length of every reverse palindrome in the string having length between 4 and 12. You may return these pairs in any order.

```python
with open('rosalind_revp.txt', 'r') as fa:
    lines = fa.readlines()
```

```python
seq_name = ''
seq_string = ''
    
for line in lines:
    line = line.strip()
    if line.startswith('>'):
        seq_name = line[1:] 
    else:
        seq_string += line 


def switch(s):
    s = s[::-1]
    switch_s = ''
    for i in range(len(s)):
        if s[i] == 'A':
            switch_s += 'T'
        elif s[i] == 'T':
            switch_s += 'A'
        elif s[i] == 'G':
            switch_s += 'C'
        elif s[i] == 'C':
            switch_s += 'G'
    return switch_s

def palindrome(s):
    for i in range(len(s)):
        for j in range(4, 13, 1):
            if s[i:i+j] == switch(s[i:i+j]) and (i+j <= len(s)):
                print(i + 1, j)

palindrome(seq_string)
```
